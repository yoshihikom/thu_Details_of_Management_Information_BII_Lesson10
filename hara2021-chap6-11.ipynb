```python
# wordcloud
# pip install wordcloud
import numpy as np
import wordcloud, codecs
import MeCab
import matplotlib.pyplot as plt

file = codecs.open('./wiki_sony_wakati.txt', 'r', 'utf8', 'ignore')
text = file.read()

m = MeCab.Tagger('')

text = text.replace("\r\n", "")
parsed = m.parse(text)

splitted = ''.join([x.split('\t')[0] for x in parsed.splitlines()[:-1] if x.split('\t')[4] not in ['-','-','']])

stop_words = [u'-',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'in',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'',u'co',u'https',u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u"",u""]
# 
font_path = "Kosugi-Regular.ttf"

wordc = wordcloud.WordCloud(font_path=font_path,background_color='white',contour_color='steelblue',contour_width=2,stopwords=set(stop_words),collocations=False,width=800,height=600).generate(splitted)
wordc.to_file('711-wordCloud-jpn.png')

plt.figure(figsize=(15, 10))
plt.imshow(wordc, interpolation="bilinear")
plt.axis("off")

#plt.savefig("result/SR.png")
plt.show()
```
