```python
# wordcloudのインストールが必要。
# pip install wordcloud
import numpy as np
import wordcloud, codecs
import MeCab
import matplotlib.pyplot as plt

file = codecs.open('./wiki_sony_wakati.txt', 'r', 'utf8', 'ignore')
text = file.read()

m = MeCab.Tagger('')

text = text.replace("\r\n", "")
parsed = m.parse(text)

splitted = '　'.join([x.split('\t')[0] for x in parsed.splitlines()[:-1] if x.split('\t')[4] not in ['助詞-格助詞','助詞-係助詞','助動詞']])

stop_words = [u'-',u'し',u'ー',u'月',u'さ',u'て',u'年',u'',u'なる',u'でき',u'てる',u'する',u'よう',u'in',u'そう',u'思っ',u'光宏',u'モデル',u'ない',u'なり',u'こと',u'なっ',u'さん',u'ください',u'co',u'https',u"あり",u"おり",u"いる",u"なら",u"する",u"ある",u"ない",u"こと",u"なっ",u"でき",u"できる",u"これ",u"もの",u"よう",u"なる",u"はず",u"なり",u"それ",u"れる",u"られる",u"ここ",u"ため",u""]
# 日本語フォントを指定
font_path = "Kosugi-Regular.ttf"

wordc = wordcloud.WordCloud(font_path=font_path,background_color='white',contour_color='steelblue',contour_width=2,stopwords=set(stop_words),collocations=False,width=800,height=600).generate(splitted)
wordc.to_file('711-wordCloud-jpn.png')

plt.figure(figsize=(15, 10))
plt.imshow(wordc, interpolation="bilinear")
plt.axis("off")

#plt.savefig("result/SR.png")
plt.show()
```
