```python
# TwitterがXになり Twitter apiが利用できなくなったため、
# Twitterから投稿を取得することはやめて、
# Wikipediaからの記事の取得とする。

# wikipediaの記事の読み取り
# 実行には，事前にwikipedia のインストールが必要
# pip install wikipedia

import codecs
import wikipedia
wikipedia.set_lang("ja")

query = "ソニー"
content = wikipedia.page(query,auto_suggest=False).content

with codecs.open('./wiki_sony.txt', 'w', 'utf8') as f:
    f.writelines(content)
f.close()
```

```python
# mecabのインストールが必要。
# Anaconda Promptで以下を実行
# pip install mecab-python3
# pip install unidic-lite

import MeCab
import codecs

tagger = MeCab.Tagger('-O wakati')

file_in = codecs.open('wiki_sony.txt', 'r', 'utf8', 'ignore')
file_out = codecs.open('wiki_sony_wakati.txt', 'w', 'utf8', 'ignore')

line = file_in.readline()

while line:
    tagger.parse("")
    result = tagger.parse("　"+line)
    file_out.write(result[1:])
    line = file_in.readline()
    print(result)

file_in.close()
file_out.close()
```
